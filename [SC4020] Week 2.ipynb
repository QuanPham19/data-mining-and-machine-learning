```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Q1: K-means on 6 points, 2 iterations, 2 clusters. A = (0,-1), B = (-2,0), C = (1,2), D = (2,1), E = (1,-1), F = (-1,2). 
# Initial cluster centroids at E and F.

data = np.array([[0,-1],[-2,0],[1,2],[2,1],[1,-1],[-1,2]])
n_clusters = 2
init_centroids = np.array([[1,-1],[-1,2]])

kmeans = KMeans(n_clusters=n_clusters, init=init_centroids, max_iter=2, n_init=1)
kmeans.fit(data)
assignments = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(data[:,0], data[:,1], c=assignments, cmap='rainbow')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', marker='x', s=100)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering (starts at E and F)')
plt.show()
```

```python
# Q2: K-means on 6 points, 2 iterations, 2 clusters. A = (0,-1), B = (-2,0), C = (1,2), D = (2,1), E = (1,-1), F = (-1,2). 
# Initial cluster centroids at C and A.

data = np.array([[0,-1],[-2,0],[1,2],[2,1],[1,-1],[-1,2]])
n_clusters = 2
init_centroids = np.array([[1,2],[0,-1]])

kmeans = KMeans(n_clusters=n_clusters, init=init_centroids, max_iter=2, n_init=1)
kmeans.fit(data)
assignments = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(data[:,0], data[:,1], c=assignments, cmap='rainbow')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', marker='x', s=100)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering (starts at A and C)')
plt.show()
```

```python
# Q3: Given the 6 data points in Q1, perform hierarchical clustering (agglomerative) until there is only 1 cluster.
# Use single linkage method.

data = np.array([[0,-1],[-2,0],[1,2],[2,1],[1,-1],[-1,2]])
link = linkage(data, method='single')

plt.figure(figsize=(10, 5))
dendrogram(link)
plt.title('Hierarchical Clustering Dendrogram (Single Linkage)')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
