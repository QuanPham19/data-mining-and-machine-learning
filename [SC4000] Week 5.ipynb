```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
```

```python
data = {'A': ['M','F','T','M','M','F','N','N','T','T'],
        'B': ['F','T','T','F','F','F','F','T','T','F'],
        'Class Label': ['+','+','+','-','-','-','-','-','-','-']}
df = pd.DataFrame(data)
df_encoded = pd.get_dummies(df, columns=['A', 'B'])
X = df_encoded.drop('Class Label', axis=1)
y = df_encoded['Class Label']

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the classifier
clf.fit(X,y)
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True, rounded=True)
plt.show()
```
